```python
#set frame number as index
df_home.set_index("Frame_number", inplace=True)
df_away.set_index("Frame_number", inplace=True)
```

```python
#import relevant packages
import pandas as pd
import numpy as np
import scipy.signal as signal
```

```python
#only speed is given in the match file. no direction can be calculated. However, this is needed for pitch control.
#Therefore, I have calculated player x, y and total speed and added it to the dataframe
#The same as Friends of tracking (youtube channel), I used a smoothing filter (Savitzky-Golay) to have more realistic calculations of the speed
def add_velocities_to_dataframe(homedf, awaydf, max_speed, smoothing=True, filter_='Savitzky-Golay', window=7, polyorder=1):
    #time between two frames
    dt = homedf['Time_in_game'].diff()
    # index of first frame in second half
    second_half_idx = homedf.Half.idxmax(2)
    
    
    # estimate velocities for players in team
    #for each player, calculate the x and y speed
    #This will give unsmoothed player velocities
    for i in range(1,12): 
        x_value = homedf['p' + str(i) +"_x"]
        x_value = pd.to_numeric(x_value, errors='coerce').astype('float64')
        y_value = homedf['p' + str(i) +"_y"]
        y_value = pd.to_numeric(y_value, errors='coerce').astype('float64')
        vx = (x_value.diff() / dt)/100
        vy = (y_value.diff() / dt)/100

        #Remove any outliers that exceed the maximum speed. These are probably position errors in the data
        if max_speed>0:
            def_speed = np.sqrt( vx**2 + vy**2 )
            vx[ def_speed>max_speed ] = np.nan
            vy[ def_speed>max_speed ] = np.nan
        vx.dropna(inplace=True)
        vy.dropna(inplace=True)
        if smoothing:
            if filter_=='Savitzky-Golay':
                # calculate first half velocity for a player
                vx.loc[:second_half_idx] = signal.savgol_filter(vx.loc[:second_half_idx],window_length=window,polyorder=polyorder)
                vy.loc[:second_half_idx] = signal.savgol_filter(vy.loc[:second_half_idx],window_length=window,polyorder=polyorder)        
                # calculate second half velocity for a player
                vx.loc[second_half_idx:] = signal.savgol_filter(vx.loc[second_half_idx:],window_length=window,polyorder=polyorder)
                vy.loc[second_half_idx:] = signal.savgol_filter(vy.loc[second_half_idx:],window_length=window,polyorder=polyorder)
       
        #add new velocities to the home dataframe
        homedf['p' + str(i) + "_vx"] = vx
        homedf['p' + str(i) + "_vy"] = vy
        homedf['p' + str(i) + '_speed'] = np.sqrt(vx**2 + vy**2)
        
        

    for i in range(1,12): # cycle through players individually
        # difference player positions in timestep dt to get unsmoothed estimate of velicity
        x_value = awaydf['p' + str(i) +"_x"]
        x_value = pd.to_numeric(x_value, errors='coerce').astype('float64')
        y_value = awaydf['p' + str(i) +"_y"]
        y_value = pd.to_numeric(y_value, errors='coerce').astype('float64')

        
        
        if max_speed>0:
            # remove unsmoothed data points that exceed the maximum speed (these are most likely position errors)
            def_speed = np.sqrt( vx**2 + vy**2 )
            vx[ def_speed>max_speed ] = np.nan
            vy[ def_speed>max_speed ] = np.nan
        vx.dropna(inplace=True)
        vy.dropna(inplace=True)
        if smoothing:
            if filter_=='Savitzky-Golay':
                # calculate first half velocity for a player
                vx.loc[:second_half_idx] = signal.savgol_filter(vx.loc[:second_half_idx],window_length=window,polyorder=polyorder)
                vy.loc[:second_half_idx] = signal.savgol_filter(vy.loc[:second_half_idx],window_length=window,polyorder=polyorder)        
                # calculate second half velocity for a player
                vx.loc[second_half_idx:] = signal.savgol_filter(vx.loc[second_half_idx:],window_length=window,polyorder=polyorder)
                vy.loc[second_half_idx:] = signal.savgol_filter(vy.loc[second_half_idx:],window_length=window,polyorder=polyorder)
        
        #add the data to the away dataframe
        awaydf['p' + str(i) + "_vx"] = vx
        awaydf['p' + str(i) + "_vy"] = vy
        awaydf['p' + str(i) + '_speed'] = np.sqrt(vx**2 + vy**2)
        
    return homedf, awaydf
```

```python
#add velocities to the dataframe by executing the function
df_home, df_away = add_velocities_to_dataframe(df_home, df_away, max_speed = 10, smoothing = True)
```

```python
#code for adding each player's maximum speed to the dataframe
#create a list of all players for each team
shirt_numbers_home = []
shirt_numbers_away = []
for j in range(1,12):
    shirt_numbers_home.extend(df_home['p' + str(j) + '_shirt_number'].unique().tolist())
    shirt_numbers_away.extend(df_away['p' + str(j) + '_shirt_number'].unique().tolist())

import numpy as np
shirt_numbers_home = np.array(shirt_numbers_home)
shirt_numbers_home = np.unique(shirt_numbers_home)
shirt_numbers_away = np.array(shirt_numbers_away)
shirt_numbers_away = np.unique(shirt_numbers_away)

df_home.fillna(0, inplace=True)
max_speed_home = []

#for each of the players, create a list with all of their speeds over the whole match and take the maximum of it
for i in shirt_numbers_home:
    speed = []
    for j in range(1,12):
        speed.extend(df_home['p' + str(j) + '_speed'].loc[df_home['p' + str(j) + '_shirt_number'] == str(i)])
    max_speed_home.append(max(speed))
    
df_away.fillna(0, inplace=True)
max_speed_away = []
for i in shirt_numbers_away:
    speed = []
    for j in range(1,12):
        speed.extend(df_away['p' + str(j) + '_speed'][df_away['p' + str(j) + '_shirt_number'] == str(i)])
    max_speed_away.append(max(speed))

#add each player's maximum speed to the original dataframe
speed = []
for i in range(len(df_home)):
    speed.append(0)
for j in range(1,12):
    df_home['p' + str(j) + '_max_speed'] = speed
    df_away['p' + str(j) + '_max_speed'] = speed
    
for i in range(len(shirt_numbers_home)):
    for j in range(1,12):
        df_home['p' + str(j) + '_max_speed'][df_home['p' + str(j) + '_shirt_number'] == shirt_numbers_home[i]] = max_speed_home[i]
for i in range(len(shirt_numbers_away)):
    for j in range(1,12):
        df_away['p' + str(j) + '_max_speed'][df_away['p' + str(j) + '_shirt_number'] == shirt_numbers_away[i]] = max_speed_away[i]
```
