```python
#find the shirt numbers of the goalkeepers for playing direction and offside purposes
def find_goalkeeper(hometeam, awayteam):
    hometeam = hometeam.iloc[0]
    awayteam = awayteam.iloc[0]
    home_players = []
    away_players = []
    GK_id = []
    #For both teams, find the player with the highest abslute x-value at kick-off. This will be the keeper
    #GK_id will return the shirt numbers of the keepers
    for i in range(1,12):
        hometeam['p'+ str(i) + '_x'] = hometeam['p'+ str(i) + '_x'].astype(np.float64)
        home_players.append(abs(hometeam['p'+ str(i) + '_x']))        
    GK_x = max(home_players)
    for i in range(1,12):
        if abs(hometeam['p'+ str(i) + '_x']) == (GK_x):
            GK_id.append(hometeam['p'+ str(i) + '_shirt_number'])
    for i in range(1,12):
        awayteam['p'+ str(i) + '_x'] = awayteam['p'+ str(i) + '_x'].astype(np.float64)
        away_players.append(abs(awayteam['p'+ str(i) + '_x']))
    GK_x = max(away_players)
    for i in range(1,12):
        if abs(awayteam['p'+ str(i) + '_x']) == (GK_x):
            GK_id.append(awayteam['p'+ str(i) + '_shirt_number'])
    return GK_id
```

```python
import pandas as pd
import numpy as np
import torch
from torch.nn.functional import softplus
from progressbar import ProgressBar
pbar = ProgressBar()

device = 'cpu'; dtype = torch.float32

#create arrays with the home positions (x and y), max speed of the home players.
home_pos = np.array([np.asarray(df_home.iloc[:,range(6 + j*6, 6 + j*6 + 2)]) for j in range(11)]) * np.array([105,68])
max_speed_home = np.array([np.asarray(df_home.iloc[:,96 + j]) for j in range(11)])
#change nan values of the home positions to very large negatvie numbers to leave players which aren't involved in the game out of the pitch control calculations
np.nan_to_num(home_pos,copy=False,nan=-1000)

#create arrays with the away positions (x and y), max speed of the away players.
away_pos = np.array([np.asarray(df_away.iloc[:,range(6 + j*6, 6 + j*6 + 2)]) for j in range(11)]) * np.array([105,68])
max_speed_away = np.array([np.asarray(df_away.iloc[:,96 + j]) for j in range(11)])
#change nan values of the away positions to very large negatvie numbers to leave players which aren't involved in the game out of the pitch control calculations
np.nan_to_num(away_pos,copy=False,nan=-1000)

#create arrays for the ball positions (x and y), time in the match, home speed (x and y), away speed (x and y), and the team in possession
ball_pos = ((np.asarray(df_home.iloc[:,range(69,71)]) * np.array([105,68])))[:,None,None,:]
tt = np.asarray(df_home['Time_in_game'])
home_v = np.array([np.asarray(df_home.iloc[:,range(74 + j*2, 74 + j*2 + 2)])for j in range(11)])
#if the speed is nan, change it to 0
np.nan_to_num(home_v,copy=False,nan=0.0)
away_v = np.array([np.asarray(df_away.iloc[:,range(74 + j*2, 74 + j*2 + 2)])for j in range(11)])
#if the speed is nan, change it to 0
np.nan_to_num(away_v,copy=False,nan=0.0)
team_in_pos = np.array(np.asarray(df_home.iloc[:, 72]))
home_pos = home_pos[:,0:,None,None,:]
away_pos = away_pos[:,0:,None,None,:]
home_v = home_v[:,:,None,None,:]
away_v = away_v[:,:,None,None,:]
ball_pos = ball_pos[None,0:]

#define important parameters
reaction_time = 0.21
average_ball_speed = 22.5
sigma = np.pi / np.sqrt(3.) / 0.45
lamb = 4.3
#define length and width of the grid for calculations
n_grid_points_x = 50
n_grid_points_y = 30
#create the grid
XX, YY = torch.meshgrid(torch.linspace(0,105,n_grid_points_x,device=device,dtype=dtype),torch.linspace(0,68,n_grid_points_y,device=device,dtype=dtype))
ti,wi = np.polynomial.legendre.leggauss(50) #used for numerical intergation later on
ti = torch.tensor(ti,device = device,dtype=dtype)
wi = torch.tensor(wi,device=device,dtype=dtype)

#all locations on the grid to calculate pitch control for
target_position = torch.stack([XX,YY],2)[None,None,:,:,:]
n_frames = len(df_home)
first_frame = 0
batch_size = 250 #to calculate pitch control for frames in batches

#create tensors for all the players
tti = torch.empty([22,batch_size,n_grid_points_x,n_grid_points_y],device = device,dtype=dtype)
tmp2 = torch.empty([22,batch_size,n_grid_points_x,n_grid_points_y,1],device = device,dtype=dtype)

#create tensor for pitch control for all the frames
pc = torch.empty([n_frames,n_grid_points_x,n_grid_points_y],device = device,dtype=dtype)


#actual pitch control calculations

#find the goalkeeper and the playing direction
GK_id = find_goalkeeper(df_home_keep, df_away_keep)
playing_direction_real = []

#iterate over all the frames in batches of 250
for f in pbar(range(int(n_frames/batch_size))):
    #create tensors for the ball positions, player positions, player speed, and player max speed for ths batch
    bp = torch.tensor(ball_pos[:,(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))],device = device,dtype=dtype)
    hp = torch.tensor(home_pos[:,(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))],device = device,dtype=dtype)
    hv = torch.tensor(home_v[:,(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))],device = device,dtype=dtype)
    ap = torch.tensor(away_pos[:,(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))],device = device,dtype=dtype)
    av = torch.tensor(away_v[:,(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))],device = device,dtype=dtype)
    msh = torch.tensor(max_speed_home[:,(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))],device = device,dtype=dtype)
    msa = torch.tensor(max_speed_away[:,(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))],device = device,dtype=dtype)
    
    #calculate the ball travel time for each location on the grid
    ball_travel_time = torch.norm(target_position - bp, dim=4).div_(average_ball_speed)
    playing_direction = []
    
    #extend the max speed tensors for calculations 
    msh = msh.unsqueeze_(-1)
    msh = msh.expand(11,250, 50)
    msh = msh.unsqueeze_(-1)
    msh = msh.expand(11,250, 50,30)
    msa = msa.unsqueeze_(-1)
    msa = msa.expand(11,250, 50)
    msa = msa.unsqueeze_(-1)
    msa = msa.expand(11,250, 50,30)
    
    #in the tensors, set the positions of the players which are offside to a very large negatvie number so that they are neglected for pitch control calculations
    #check which team is in possession
    for i in range(first_frame + f*batch_size, np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames))):
        if team_in_pos[i] == 'H':
            #find goalkeeper number
            defending_keeper = GK_id[1]
            x_positions = []
            for j in range(1,12):
                if df_away.iloc[i]['p' + str(j) + '_shirt_number'] == GK_id[1]:
                    keeper_id = 'p' + str(j)
                    keeper_x = df_away_keep.iloc[i][str(keeper_id) + '_x']
                    defending_half = np.sign(keeper_x)
                x_positions.append(df_away_keep.iloc[i]['p' + str(j) + '_x'])
            second_deepest_defender_x = sorted( [defending_half*p for p in x_positions], reverse=True )[1]
            #define the offside line
            offside_line = max(second_deepest_defender_x,defending_half*df_home_keep.iloc[i]['Ball_x_position'],0.0)+0.2
            offside_line = ((offside_line + 5250) /10500)*105
            for j in range(0,11):
                #check whether a player is ofside
                if hp[j][i-(np.minimum(first_frame + f*batch_size,int(first_frame + n_frames)))-1][0][0][0].item()*defending_half >= offside_line:
                    #set position to very large negative number if offside
                    hp[j][i-(np.minimum(first_frame + f*batch_size,int(first_frame + n_frames)))-1][0][0][0] = -1000000
                    
        else:
            #find goalkeeper number
            defending_keeper = GK_id[0]
            x_positions = []
            for j in range(1,12):
                if df_home.iloc[i]['p' + str(j) + '_shirt_number'] == GK_id[0]:
                    keeper_id = 'p' + str(j)
                    keeper_x = df_home_keep.iloc[i][str(keeper_id) + '_x']
                    defending_half = np.sign(keeper_x)
                x_positions.append(df_home_keep.iloc[i]['p' + str(j) + '_x'])
            second_deepest_defender_x = sorted( [defending_half*p for p in x_positions], reverse=True )[1]
            #define the offside line
            offside_line = max(second_deepest_defender_x,defending_half*df_home_keep.iloc[i]['Ball_x_position'],0.0)+0.2
            offside_line = ((offside_line + 5250) /10500)*105
            for j in range(0,11):
                #check if players are offside
                if ap[j][i-(np.minimum(first_frame + f*batch_size,int(first_frame + n_frames)))-1][0][0][0].item()*defending_half >= offside_line:
                    #set position to very large negative number if offside
                    ap[j][i-(np.minimum(first_frame + f*batch_size,int(first_frame + n_frames)))-1][0][0][0] = -1000000
        
        #add playing direction for PCV calculaions later on 
        if keeper_x < 0:
            playing_direction.append('Left')
        else:
            playing_direction.append('Right')
    for z in range(len(playing_direction)):
        playing_direction_real.append(playing_direction[z])
        
    #calculate the position of the players after reaction time and the distance of that position to  the target position
    r_reaction_home = hp + hv.mul_(reaction_time)
    r_reaction_away = ap + av.mul_(reaction_time)
    r_reaction_home = r_reaction_home - target_position
    r_reaction_away = r_reaction_away - target_position
    
    #calculate how long it takes every player to reach the target location
    tti[:11,:ball_travel_time.shape[1]] = torch.norm(r_reaction_home,dim=4).add_(reaction_time).div_(msh)
    tti[11:,:ball_travel_time.shape[1]] = torch.norm(r_reaction_away,dim=4).add_(reaction_time).div_(msa)
    
    #calculate the uncertainty of the ball arrival time
    tmp2[...,0] = sigma * (ball_travel_time - tti)
    tmp1 = sigma * 0.5 * (ti + 1) * 10 + tmp2
    
    #calculate the time to control of the players
    hh = torch.sigmoid(tmp1[:11]).mul_(4.3)
    h = hh.sum(0)
    S = torch.exp(-lamb*torch.sum(softplus(tmp1) - softplus(tmp2),dim=0).div_(sigma))

    #calculate pitch control at each of the locations on the grid
    pc[(first_frame + f*batch_size):(np.minimum(first_frame + (f+1)*batch_size,int(first_frame + n_frames)))] = torch.matmul(S*h,wi).mul_(5.)
```

```python
#create lists for pitch control of the home and away team separately
pc_home = pc[0:(f+1)*250]
pc_away = 1-pc[0:(f+1)*250]
```

```python
#PCV calculations
from statistics import mean
import pickle
import copy
pbar = ProgressBar()

#import xT grid
df_xT = pd.read_csv('xT.csv', header=None)
xTMap = torch.tensor(df_xT.values)
xTMap_interp = torch.nn.functional.interpolate(xTMap[None,None,:,:],(n_grid_points_x,n_grid_points_y),mode='bilinear')
#flip xT grid fro calculations when the team's are playing in the other direction
xTflipped = torch.flip(xTMap_interp[0,0],[0,1])
expected_threat = xTMap_interp[0,0].reshape((1,-1))
expected_threat_away = xTflipped.reshape((1,-1))

#initialize lists for pitch control and PCV (weighted pitch control)
pc_home_new = []
pc_away_new = []
weighted_PPCF_home_new = []
weighted_PPCF_away_new = []
weighted_PPCF_home = []
pitch_control_home = []
weighted_PPCF_away = []
pitch_control_away = []
xy_balls = torch.Tensor(ball_pos)

#iterate over the length of the playing direction list. = same length as the pitch control lists
for i in pbar(range(len(playing_direction_real))):
    xy_ball = torch.Tensor(xy_balls[0][i])
    weighted_pc_home = []
    weighted_pc_away = []
    pitch_control_home = []
    pitch_control_away = []
    xy_query = []

    #add the average pitch control for the home and away team
    for z in range(n_grid_points_x):
        for j in range(n_grid_points_y):
            pitch_control_home.append(pc_home[i][z][j].item())
            pitch_control_away.append(pc_away[i][z][j].item())
            xy_query.append([XX[z][0].item(), YY[0][j].item()])
    xy_query = torch.Tensor(xy_query)
    pc_home_new.append(mean(pitch_control_home))
    pc_away_new.append(mean(pitch_control_away))
    pitch_control_home = torch.Tensor(pitch_control_home)
    pitch_control_away = torch.Tensor(pitch_control_away)
    
    #transition layer for the probability that the next on-ball event will happen at a specific location given the ball's current location.
    transition_layer = torch.exp(( - torch.sqrt(torch.sum(((xy_query) - (xy_ball[0][0])) **2,axis = 1))/ (2*14)))
    transition_layer = transition_layer / transition_layer.sum()
    
    #multiply pitch control surface with the transition layer
    weighted_ball_pos_home = pitch_control_home*transition_layer
    weighted_ball_pos_home = torch.max(weighted_ball_pos_home,torch.zeros_like(weighted_ball_pos_home))
    weighted_ball_pos_away = pitch_control_away * transition_layer
    weighted_ball_pos_away = torch.max(weighted_ball_pos_away,torch.zeros_like(weighted_ball_pos_away))
    
    #multiply with the xT surface
    if playing_direction_real[i] == 'Left':
        weighted_pc_home.append(weighted_ball_pos_home*expected_threat_away)
        weighted_pc_away.append(weighted_ball_pos_away*expected_threat_away)
    else:
        weighted_pc_home.append(weighted_ball_pos_home*expected_threat)
        weighted_pc_away.append(weighted_ball_pos_away*expected_threat)
    
    #add the PCV values to the lists
    weighted_PPCF_home_new.append(mean(weighted_pc_home[0][0].tolist()))
    weighted_PPCF_away_new.append(mean(weighted_pc_away[0][0].tolist()))
    weighted_pc_home = np.reshape(weighted_pc_home[0][0], (n_grid_points_x, n_grid_points_y))
    weighted_PPCF_home.append(weighted_pc_home)
    weighted_pc_away = np.reshape(weighted_pc_away[0][0], (n_grid_points_x, n_grid_points_y))
    weighted_PPCF_away.append(weighted_pc_away)
```

```python
#add the average values for pitch control and PCV to the original home and away dataframes
df_home['PPCF'] = pc_home_new
df_home['Weighted_PPCF'] = weighted_PPCF_home_new
df_away['PPCF'] = pc_away_new
df_away['Weighted_PPCF'] = weighted_PPCF_away_new
```
