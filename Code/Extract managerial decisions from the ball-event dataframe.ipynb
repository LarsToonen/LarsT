```python
#function to check whether a red card is given
def check_red_card(f24_file):
    tree = et.ElementTree(file = f24_file)
    gameFile = tree.getroot()
    for game in gameFile:
        #Iterate through each event
        for event in game:
            for qualifier in event:
                if qualifier.attrib.get("qualifier_id") == "32" or qualifier.attrib.get("qualifier_id") == "33":
                    return 'True'
```

```python
#code to extract the managerial decisions
from progressbar import ProgressBar
import os
import xml.etree.ElementTree as ET
import pickle
final_df = pd.DataFrame()
df_type_of_decision_fromation = pd.read_excel('offensiveness_formations.xlsx')
path = 'C:\\Users\\s153035\\Dropbox\\Lars Toonen\\Data'
#iterate over each match
for folder1 in os.listdir(path):
    pbar = ProgressBar()
    for folder2 in pbar(os.listdir(path + '\\' + str(folder1))):
        for filename in os.listdir(path + '\\' + str(folder1) + '\\'+ str(folder2)):
            for all_files in os.listdir(path + '\\' + str(folder1) + '\\' + str(folder2)+ '\\' + str(filename)):
                if folder1 == 'Seizoen 1819' or folder1 == 'Seizoen 2021':
                    if 'f24' in all_files:
                        event_file = os.path.join(path + '\\' + str(folder1) + '\\' + str(folder2) + '\\'+ str(filename), all_files)
                    elif all_files.endswith('metadata.xml'):
                        metadata_file = os.path.join(path + '\\' + str(folder1) + '\\' + str(folder2) + '\\'+ str(filename), all_files)
                else:
                    if 'f24' in all_files:
                        event_file = os.path.join(path + '\\' + str(folder1) + '\\' + str(folder2) + '\\'+ str(filename), all_files)
                        for files in  os.listdir(path + '\\' + str(folder1) + '\\' + str(folder2)+ '\\' + str(filename) + '\\' + str(filename) + '_TracDAT'):
                            if all_files.endswith('metadata.xml'):
                                metadata_file = os.path.join(path + '\\' + str(folder1) + '\\' + str(folder2)+ '\\' + str(filename) + '\\' + str(filename) + '_TracDAT', files)
 
            #check if red card is given
            if check_red_card(event_file) == 'True':
                pass
            else:
                tree = et.ElementTree(file = event_file)
                gameFile = tree.getroot()
                team_dict = {gameFile[0].attrib["home_team_id"]: 'Home',
                gameFile[0].attrib["away_team_id"]: 'Away'}
                main_id = []
                event_id = []
                type_id = []
                half = []
                frame_id = []
                minute = []
                seconds = []
                team_id = []
                team = []
                formation_change = []
                players_substituted = []

                tree2 = et.ElementTree(file = metadata_file)
                meta_data = tree2.getroot()
                start_of_halves = []
                end_of_halves = []
                for i in meta_data:
                    for j in i:
                        start_of_halves.append(j.attrib.get("iStartFrame"))
                        end_of_halves.append(j.attrib.get("iEndFrame"))

                for game in gameFile:

                    #Iterate through each event
                    for event in game:

                        #To the correct list, append the correct attribute using attrib.get()
                        main_id.append(event.attrib.get("id"))
                        event_id.append(event.attrib.get("event_id"))
                        type_id.append(event.attrib.get("type_id"))
                        half.append(event.attrib.get("period_id"))
                        frame_id.append((int(event.attrib.get("min")) * 60 + int(event.attrib.get("sec")))*25)
                        minute.append(event.attrib.get("min"))
                        seconds.append(event.attrib.get("sec"))
                        team_id.append(event.attrib.get("team_id"))
                        team.append(team_dict[event.attrib.get("team_id")])
                        formation_change.append("")
                        players_substituted.append("")
                        
                        #add which player is substituted (defender, midfielder, or attacker)
                        for qualifier in event:
                            if qualifier.attrib.get("qualifier_id") == "44":
                                players_substituted[-1] = qualifier.attrib.get("value")
                                

                        #add the formation change (number of the new formation) 
                        for qualifier in event:
                            if qualifier.attrib.get("qualifier_id") == "130":
                                formation_change[-1] = (qualifier.attrib.get("value"))

                #Create a list of columns
                column_titles = ["main_id", "event_id", "type_id", "half", "frame_id", "minute", "seconds", "team_id", "team", "formation_change", "substituted players"]

                #Use pd.DataFrame to create our table, assign the data in the order of our columns and give it the column titles above
                final_table = pd.DataFrame(data=[main_id, event_id, type_id, half, frame_id, minute, seconds, team_id, team, formation_change, players_substituted], index=column_titles)

                #Transpose, or flip, the table. Otherwise, our table will run from left to right, rather than top to bottom
                final_table = final_table.T

                #add frame ids to the table corresponding to the frames in the tracking dataframe
                for i in range(len(final_table)):
                    if final_table.iloc[i]['half'] == '1':
                        final_table.iloc[i, final_table.columns.get_loc('frame_id')] = int(start_of_halves[0]) + int(final_table.iloc[i]['frame_id'])
                    elif final_table.iloc[i]['half'] == '2':
                        final_table.iloc[i, final_table.columns.get_loc('frame_id')] = int(start_of_halves[1]) + int(final_table.iloc[i]['frame_id']) - 45*60*25

                #add the names of the event types (same as in creating the ball-event dataframe)
                name_of_types = []
                for i in range(len(final_table)):
                    if final_table.iloc[i]['type_id'] == '1':
                        name_of_types.append("Pass")
                    elif final_table.iloc[i]['type_id'] == '2':
                        name_of_types.append("Offside pass")
                    elif final_table.iloc[i]['type_id'] == '3':
                        name_of_types.append("Attempted dribble")
                    elif final_table.iloc[i]['type_id'] == '4':
                        name_of_types.append("Foul commited")
                    elif final_table.iloc[i]['type_id'] == '5':
                        name_of_types.append("out")
                    elif final_table.iloc[i]['type_id'] == '6':
                        name_of_types.append("Corner awarded")
                    elif final_table.iloc[i]['type_id'] == '7':
                        name_of_types.append("Tackle")
                    elif final_table.iloc[i]['type_id'] == '8':
                        name_of_types.append("Interception")
                    elif final_table.iloc[i]['type_id'] == '10':
                        name_of_types.append("Goalkeeper save")
                    elif final_table.iloc[i]['type_id'] == '11':
                        name_of_types.append("Goalkeeper catches cross")
                    elif final_table.iloc[i]['type_id'] == '12':
                        name_of_types.append("Defensive clearance")
                    elif final_table.iloc[i]['type_id'] == '13':
                        name_of_types.append("Shot missed wide")
                    elif final_table.iloc[i]['type_id'] == '14':
                        name_of_types.append("Ball hits frame of the goal")
                    elif final_table.iloc[i]['type_id'] == '15':
                        name_of_types.append("Shot attempt saved")
                    elif final_table.iloc[i]['type_id'] == '16':
                        name_of_types.append("Goal")
                    elif final_table.iloc[i]['type_id'] == '17':
                        name_of_types.append("Card")
                    elif final_table.iloc[i]['type_id'] == '18':
                        name_of_types.append("Player substituted off")
                    elif final_table.iloc[i]['type_id'] == '19':
                        name_of_types.append("Player substituted on")
                    elif final_table.iloc[i]['type_id'] == '20':
                        name_of_types.append("Player retired")
                    elif final_table.iloc[i]['type_id'] == '21':
                        name_of_types.append("Player returns on pitch")
                    elif final_table.iloc[i]['type_id'] == '22':
                        name_of_types.append("Player becomes goalkeeper")
                    elif final_table.iloc[i]['type_id'] == '23':
                        name_of_types.append("Goalkeeper becomes player")
                    elif final_table.iloc[i]['type_id'] == '24':
                        name_of_types.append("Condition change")
                    elif final_table.iloc[i]['type_id'] == '25':
                        name_of_types.append("Change in official")
                    elif final_table.iloc[i]['type_id'] == '27':
                        name_of_types.append("Start Delay")
                    elif final_table.iloc[i]['type_id'] == '28':
                        name_of_types.append("End delay")
                    elif final_table.iloc[i]['type_id'] == '30':
                        name_of_types.append("End period")
                    elif final_table.iloc[i]['type_id'] == '32':
                        name_of_types.append("Start period")
                    elif final_table.iloc[i]['type_id'] == '34':
                        name_of_types.append("Team setup")
                    elif final_table.iloc[i]['type_id'] == '35':
                        name_of_types.append("Player changed position")
                    elif final_table.iloc[i]['type_id'] == '36':
                        name_of_types.append("Player changed jersey number")
                    elif final_table.iloc[i]['type_id'] == '37':
                        name_of_types.append("Data collection end")
                    elif final_table.iloc[i]['type_id'] == '38':
                        name_of_types.append("Temporary goal")
                    elif final_table.iloc[i]['type_id'] == '39':
                        name_of_types.append("Temporary attempt")
                    elif final_table.iloc[i]['type_id'] == '40':
                        name_of_types.append("Formation change")
                    elif final_table.iloc[i]['type_id'] == '41':
                        name_of_types.append("Goalkeeper punch")
                    elif final_table.iloc[i]['type_id'] == '42':
                        name_of_types.append("Good player skill")
                    elif final_table.iloc[i]['type_id'] == '43':
                        name_of_types.append("Deleted event")
                    elif final_table.iloc[i]['type_id'] == '44':
                        name_of_types.append("Aerial duel")
                    elif final_table.iloc[i]['type_id'] == '45':
                        name_of_types.append("Unsuccessful tackle")
                    elif final_table.iloc[i]['type_id'] == '49':
                        name_of_types.append("Ball recovery")
                    elif final_table.iloc[i]['type_id'] == '50':
                        name_of_types.append("Dispossessed")
                    elif final_table.iloc[i]['type_id'] == '51':
                        name_of_types.append("Error leading to shot or goal")
                    elif final_table.iloc[i]['type_id'] == '52':
                        name_of_types.append("Goalkeeper picks up the ball")
                    elif final_table.iloc[i]['type_id'] == '53':
                        name_of_types.append("Goalkeeper fails to interrupt cross")
                    elif final_table.iloc[i]['type_id'] == '54':
                        name_of_types.append("Goalkeeper covers a ground ball")
                    elif final_table.iloc[i]['type_id'] == '55':
                        name_of_types.append("Offside provoked")
                    elif final_table.iloc[i]['type_id'] == '56':
                        name_of_types.append("Shield ball that goes out of play")
                    elif final_table.iloc[i]['type_id'] == '57':
                        name_of_types.append("Incorrect throw-in")
                    elif final_table.iloc[i]['type_id'] == '58':
                        name_of_types.append("Goalkeeper faces penalty")
                    elif final_table.iloc[i]['type_id'] == '59':
                        name_of_types.append("Goalkeeper comes off the line to clear or claim the ball")
                    elif final_table.iloc[i]['type_id'] == '60':
                        name_of_types.append("Chance missed (no shot)")
                    elif final_table.iloc[i]['type_id'] == '61':
                        name_of_types.append("Bad touch and lost possession")
                    elif final_table.iloc[i]['type_id'] == '63':
                        name_of_types.append("Temporary save")
                    elif final_table.iloc[i]['type_id'] == '64':
                        name_of_types.append("Match resumes after the game was abandoned")
                    elif final_table.iloc[i]['type_id'] == '65':
                        name_of_types.append("Contentious referee decision")
                    elif final_table.iloc[i]['type_id'] == '67':
                        name_of_types.append("2 players running for a loose ball")
                    elif final_table.iloc[i]['type_id'] == '68':
                        name_of_types.append("Referee drop ball")
                    elif final_table.iloc[i]['type_id'] == '70':
                        name_of_types.append("Injury time announcement")
                    elif final_table.iloc[i]['type_id'] == '71':
                        name_of_types.append("Coach setup")
                    elif final_table.iloc[i]['type_id'] == '74':
                        name_of_types.append("Blocked pass")
                    elif final_table.iloc[i]['type_id'] == '75':
                        name_of_types.append("Delayed start")
                    elif final_table.iloc[i]['type_id'] == '76':
                        name_of_types.append("Early end")
                    elif final_table.iloc[i]['type_id'] == '77':
                        name_of_types.append("Player off pitch")
                    elif final_table.iloc[i]['type_id'] == '79':
                        name_of_types.append("Coverage interruption")
                    elif final_table.iloc[i]['type_id'] == '80':
                        name_of_types.append("Goalkeeper drops ball")
                    elif final_table.iloc[i]['type_id'] == '81':
                        name_of_types.append("Ball touches a non-player object")
                    elif final_table.iloc[i]['type_id'] == '83':
                        name_of_types.append("Unsuccessful tackle")
                final_table.insert(3, 'type_name', name_of_types)
                
                #create final table with only decision events
                to_include = ['16', '18','19', '34','35', '40']
                final_table = final_table.loc[final_table['type_id'].isin(to_include)]
                final_table = final_table.reset_index()
                home_score = []
                away_score = []
                minutes_left = []
                seconds_left = []
                home = 0
                away = 0
                #add the home and away score to the table at the moment of the decision
                for i in range(len(final_table)):
                    minutes_left.append(90 - int(final_table.iloc[i]['minute']))
                    seconds_left.append(60 - int(final_table.iloc[i]['seconds']))
                    if final_table.iloc[i]['type_id'] == '16':
                        if final_table.iloc[i]['team'] == 'Home':
                            home = home + 1
                            home_score.append(home)
                            away_score.append(away)
                        else:
                            away = away + 1
                            away_score.append(away)
                            home_score.append(home)
                    else:
                        home_score.append(home)
                        away_score.append(away)
                final_table['home score'] = home_score
                final_table['away score'] = away_score
                final_table['minutes left'] = minutes_left
                final_table['seconds left'] = seconds_left
                to_include = ['18', '19', '34','35', '40']
                final_table = final_table.loc[final_table['type_id'].isin(to_include)]
                final_table = final_table.reset_index()
                current_formation_home = []
                current_formation_away = []
                #add the current formation of the teams to the dataframe
                for i in range(len(final_table)):
                    if final_table.iloc[i]['type_id'] == '34':
                        if final_table.iloc[i]['team'] == 'Away':
                            formation_away = final_table.iloc[i]['formation_change']
                            current_formation_away.append(formation_away)
                            if len(current_formation_home) == 0:
                                formation_home  = 0
                                current_formation_home.append(formation_home)
                            else:
                                current_formation_home.append(formation_home)
                        else:
                            formation_home = final_table.iloc[i]['formation_change']
                            current_formation_home.append(formation_home)
                            if len(current_formation_away) == 0:
                                formation_away  = 0
                                current_formation_away.append(formation_away)
                            else:
                                current_formation_away.append(formation_away)
                    elif final_table['type_id'][i] == '40':
                        if final_table['team'][i] == 'Away':
                            formation_away = final_table.iloc[i]['formation_change']
                            current_formation_away.append(formation_away)
                            current_formation_home.append(formation_home)
                        else:
                            formation_home = final_table.iloc[i]['formation_change']
                            current_formation_home.append(formation_home)
                            current_formation_away.append(formation_away)
                    else:
                        current_formation_home.append(formation_home)
                        current_formation_away.append(formation_away)

                final_table['current_formation_home'] = current_formation_home
                final_table['current_formation_away'] = current_formation_away
                
                decision = []
                #add the type of decisions to the dataframe (either offensive, neutral, or defensive)
                for i in range(len(final_table)):
                    #if team setup: neutral
                    if final_table.iloc[i]['type_id'] == '34':
                        decision.append('neutral')
                    #if player changed position: neutral
                    elif final_table.iloc[i]['type_id'] == '35':
                        decision.append('neutral')
                    #if formation change: decision depending on the formation change
                    elif final_table.iloc[i]['type_id'] == '40':
                        if final_table.iloc[i]['team'] == 'Home':
                            decision.append(df_type_of_decision_fromation.iloc[int(final_table.iloc[i]['formation_change'])-1][int(final_table.iloc[i-1]['current_formation_home'])-1])
                        if final_table.iloc[i]['team'] == 'Away':
                            decision.append(df_type_of_decision_fromation.iloc[int(final_table.iloc[i]['formation_change'])-1][int(final_table.iloc[i-1]['current_formation_away'])-1])
                    #if substition: depending on the substitution
                    elif final_table.iloc[i]['type_id'] == '19':
                        try:
                            if int(final_table['minute'][final_table.loc[final_table['type_id'] == '40'].index[final_table.loc[final_table['type_id'] == '40'].index >= i].min()]) - int(final_table.iloc[i]['minute']) <= 2:
                                decision.append('neutral')
                            else:
                                decision.append('neutral')
                        except:
                            decision.append('neutral')
                    else:
                        decision.append('neutral')
                substitutions = []
                for i in range(len(final_table)):
                    if final_table.iloc[i]['type_id'] == '19':
                        substitutions.append([final_table.iloc[i]['substituted players'], final_table.iloc[i-1]['substituted players'] ])
                    else:
                        substitutions.append([])
                
                #create final dataframe
                final_table['players involved'] = substitutions
                final_table['Type of decision'] = decision
                to_include = to_include = ['19', '35', '40']
                final_table = final_table.loc[final_table['type_id'].isin(to_include)]
                final_table = final_table.reset_index(drop=True)
                index_to_drop = []
                final_table_home = final_table.loc[final_table['team'] == 'Home']
                final_table_away = final_table.loc[final_table['team'] == 'Away']
                
                #if substitutionn is directly followed by a formation change, drop the substitution from the dataframe
                for i in final_table_home.index:
                    if final_table_home['type_id'][i] == '19':
                        try:
                            if int(final_table_home['minute'][final_table_home.loc[final_table_home['type_id'] == '40'].index[final_table_home.loc[final_table_home['type_id'] == '40'].index >= i].min()]) - int(final_table_home['minute'][i]) <= 2:
                                index_to_drop.append(i)
                        except:
                            pass
                for i in final_table_away.index:
                    if final_table_away['type_id'][i] == '19':
                        try:
                            if int(final_table_away['minute'][final_table_away.loc[final_table_away['type_id'] == '40'].index[final_table_away.loc[final_table_away['type_id'] == '40'].index >= i].min()]) - int(final_table_away['minute'][i]) <= 2:
                                index_to_drop.append(i)
                        except:
                            pass
                final_table = final_table.drop(final_table.index[index_to_drop])
                final_table = final_table.reset_index(drop=True)
                match_id = []
                index_to_drop = []
                #drop team setup events
                for i in range(len(final_table)):
                    match_id.append(filename)
                    if final_table.iloc[i]['type_id'] == '40' and final_table.iloc[i]['minute'] == '0':
                        index_to_drop.append(i)
                        match_id.pop(-1)
                final_table = final_table.drop(final_table.index[index_to_drop])
                final_table = final_table.reset_index(drop=True)
                final_table['Match'] = match_id
                final_table = final_table.drop(columns=['substituted players'])
                final_df = pd.concat([final_df, final_table])


                
                
```

```python
#in all decisions, the type of decision is adjusted based on some conditions, these conditions are documented in the report
for i in range(len(final_df)):
    try:
        if final_df.iloc[i]['players involved'][0] == 'Midfielder' and final_df.iloc[i]['players involved'][1] == 'Defender':
            final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'offensive'
        elif final_df.iloc[i]['players involved'][0] == 'Forward' and final_df.iloc[i]['players involved'][1] == 'Midfielder':
            final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'offensive'
        elif final_df.iloc[i]['players involved'][0] == 'Forward' and final_df.iloc[i]['players involved'][1] == 'Defender':
            final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'offensive'
        elif final_df.iloc[i]['players involved'][0] == 'Defender' and final_df.iloc[i]['players involved'][1] == 'Midfielder':
            final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'defensive'
        elif final_df.iloc[i]['players involved'][0] == 'Midfielder' and final_df.iloc[i]['players involved'][1] == 'Forward':
            final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'defensive'
        elif final_df.iloc[i]['players involved'][0] == 'Defender' and final_df.iloc[i]['players involved'][1] == 'Forward':
            final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'defensive'
        elif final_df.iloc[i]['players involved'][0] == 'Forward' and final_df.iloc[i]['players involved'][1] == 'Forward':
            if final_df.iloc[i]['team'] == 'Home':
                if final_df.iloc[i]['home score'] - final_df.iloc[i]['away score'] <0:
                    final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'offensive'
            else:
                if final_df.iloc[i]['away score'] - final_df.iloc[i]['home score'] <0:
                    final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'offensive'  
        elif final_df.iloc[i]['players involved'][0] == 'Defender' and final_df.iloc[i]['players involved'][1] == 'Defender':
            if final_df.iloc[i]['team'] == 'Home':
                if final_df.iloc[i]['home score'] - final_df.iloc[i]['away score'] >0:
                    final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'defensive'
            else:
                if final_df.iloc[i]['away score'] - final_df.iloc[i]['home score'] >0:
                    final_df.iloc[i, final_df.columns.get_loc('Type of decision')] = 'defensive'   
    except:
        pass
                
```
