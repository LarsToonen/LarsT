```python
final_final_table = final_final_table.reset_index()
```

```python
#for dribbles, add the destination of the dribble
df_dribbles = final_final_table.loc[final_final_table['type_id'] == '3']
for i in df_dribbles['index']:
    try:
        final_final_table['x_destination'][i] = final_final_table['x_origin'][i+1]
        final_final_table['y_destination'][i] = final_final_table['y_origin'][i+1]
    except:
        final_final_table['x_destination'][i] = 0
        final_final_table['y_destination'][i] = 0
```

```python
#only inlcude events that are needed for calculation of xT
to_include = ['1', '3', '13', '15', '16']
df_events = final_final_table.loc[final_final_table.type_id.isin(to_include)]
df_events = df_events.rename(columns = {'x_origin': 'start_x', 'y_origin': 'start_y', 'x_destination': 'end_x', 'y_destination': 'end_y'}, inplace = False)
df_events = df_events.reset_index(drop = True)
```

```python
#outcome is 1 only if a goal is made. Used in calculations
for i in range(len(df_events)):
    if df_events['type_id'][i] == '13':
        df_events['type_name'][i] = 'shot'
        df_events['outcome'][i] = '0'
    elif df_events['type_id'][i] == '15':
        df_events['type_name'][i] = 'shot'
        df_events['outcome'][i] = '0'
    elif df_events['type_id'][i] == '16':
        df_events['type_name'][i] = 'shot'
        df_events['outcome'][i] = '1'
```

```python
#convert start location to a float
import numpy as np
df_events['start_x'] = df_events['start_x'].astype(np.float64)
df_events['start_y'] = df_events['start_y'].astype(np.float64)
```

```python
#if event has no end destination: add 0 as end destination
for i in range(len(df_events)):
    if df_events['end_x'][i] == '':
        df_events['end_x'][i] = 0
        df_events['end_y'][i] = 0
        
```

```python
#convert the end destination of the event to a float
df_events['end_x'] = df_events['end_x'].astype(np.float64)
df_events['end_y'] = df_events['end_y'].astype(np.float64)
```

```python
#adjust for the pitch boundaries
df_events['start_x'] = df_events['start_x'] * 1.05
df_events['start_y'] = df_events['start_y'] * 0.68
df_events['end_x'] = df_events['end_x'] * 1.05
df_events['end_y'] = df_events['end_y'] * 0.68
```

```python
#create the xT surface
xTModel = xthreat.ExpectedThreat(l=50, w=50)
xTModel.fit(df_events)
```
